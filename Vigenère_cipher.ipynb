```python
import re
import hashlib
import string
import random

def vig_encrypt(txt : str, key : str) -> str:
  def vig_init(vig_key : str) -> str:
    vig_txt = 'abcdefghijklmnopqrstuvwxyz'
    vig_txt = vig_txt.upper()
    vig_key = vig_key.upper()

    for i in vig_key:
      if i in vig_txt:
        vig_txt = vig_txt.replace(i, '')
    vig = vig_key + vig_txt
    return vig

  def vig_table_init(vig_key : str) -> dict:
    vig = vig_init(vig_key)
    row_index = '_' + vig
    vig_table = [vig[i:] + vig[:i] for i in range(len(vig))]
    vig_table = [row_index[1:]] + vig_table
    vig_table = dict(zip(row_index, vig_table))
    return vig_table

  def process_key(key : str) -> str:
    #h = SHA256_digest(key)
    h = hashlib.sha256(key.encode()).hexdigest()
    hl = [h[i:i+8] for i in range(0,64,8)] #647e3264 10ba83a9 91aef4d5 49385719 1d5c1f80 2c33e62a 9a7e49eb 9a02a2ab
    ktable = hl.pop(len(key)%8)
    kenc = ''.join(''.join(chr((int(c, 16) % 26) + ord('A')) for c in hl[i]) for i in range(len(hl)))
    ktable = ''.join(chr((int(c, 16) % 26) + ord('A')) for c in ktable)
    used = set()
    alphabet = iter(ch for ch in string.ascii_uppercase if ch not in ktable)
    result = []
    for ch in ktable:
        if ch in used:
            replacement = next(alphabet)
            result.append(replacement)
            used.add(replacement)
        else:
            result.append(ch)
            used.add(ch)
    #print("".join(result), kenc)
    return "".join(result), kenc

  s = "".join(re.sub(r"\s+", "", x) for x in re.split(r"[,.?!]", txt.upper()))
  key = "".join(re.sub(r"\s+", "", x) for x in re.split(r"[,.?!]", key.upper()))
  k1, k2 = process_key(key)

  vig_table = vig_table_init(k1)
  n = len(s)//len(k2) + 1
  key_txt = k2*n
  key_txt = key_txt[:len(s)]

  c = ''
  for i, j in (zip(s, key_txt)):
    c += vig_table[i][vig_table['_'].find(j)]

  s1 = txt
  s2 = c

  core = []
  j = 0
  space_count = s1.count(" ")
  for ch in s1:
      if ch.isalpha():
          core.append(s2[j])
          j += 1
      elif ch in ".,?!":
          core.append(ch)

  text_no_space = "".join(core)

  positions = list(range(1, len(text_no_space)))
  chosen = set()
  while len(chosen) < space_count:
      pos = random.choice(positions)
      if pos in chosen or pos - 1 in chosen or pos + 1 in chosen:
          continue
      chosen.add(pos)

  result = []
  for i, ch in enumerate(text_no_space):
      if i in chosen:
          result.append(" ")
      result.append(ch)

  c_new = "".join(result)

  return c_new

print(vig_encrypt('Am I happy?','Smiling'))
```

```python
def SHA256_digest(s : str) -> str:
  msg_bytes = [format(ord(c), '08b') for c in s]

  message_bit_len = len(msg_bytes) * 8
  k = (448 - (message_bit_len + 1)) % 512
  pad_bits = '1' + ('0' * k)
  pad_bytes = [pad_bits[i:i+8] for i in range(0, len(pad_bits), 8)]

  len64 = format(message_bit_len, '064b')
  len_bytes = [len64[i:i+8] for i in range(0, 64, 8)]

  m_bytes = msg_bytes + pad_bytes + len_bytes
  s_512 = ''.join(m_bytes)
  blocks = [s_512[i:i+512] for i in range(0, len(s_512), 512)]
  words32 = [[block[i:i+32] for i in range(0, 512, 32)] for block in blocks]


  K = [0x428a2f98, 0x71374491, 0xb5c0fbcf, 0xe9b5dba5, 0x3956c25b, 0x59f111f1, 0x923f82a4, 0xab1c5ed5,
      0xd807aa98, 0x12835b01, 0x243185be, 0x550c7dc3, 0x72be5d74, 0x80deb1fe, 0x9bdc06a7, 0xc19bf174,
      0xe49b69c1, 0xefbe4786, 0x0fc19dc6, 0x240ca1cc, 0x2de92c6f, 0x4a7484aa, 0x5cb0a9dc, 0x76f988da,
      0x983e5152, 0xa831c66d, 0xb00327c8, 0xbf597fc7, 0xc6e00bf3, 0xd5a79147, 0x06ca6351, 0x14292967,
      0x27b70a85, 0x2e1b2138, 0x4d2c6dfc, 0x53380d13, 0x650a7354, 0x766a0abb, 0x81c2c92e, 0x92722c85,
      0xa2bfe8a1, 0xa81a664b, 0xc24b8b70, 0xc76c51a3, 0xd192e819, 0xd6990624, 0xf40e3585, 0x106aa070,
      0x19a4c116, 0x1e376c08, 0x2748774c, 0x34b0bcb5, 0x391c0cb3, 0x4ed8aa4a, 0x5b9cca4f, 0x682e6ff3,
      0x748f82ee, 0x78a5636f, 0x84c87814, 0x8cc70208, 0x90befffa, 0xa4506ceb, 0xbef9a3f7, 0xc67178f2] #1st 64 decimal places of cube root of 1st 64 prime numbers.
  K = [format(k, '032b') for k in K]

  H = [0x6a09e667, 0xbb67ae85, 0x3c6ef372, 0xa54ff53a, 0x510e527f, 0x9b05688c, 0x1f83d9ab, 0x5be0cd19] #1st 64 decimal places of square root of 1st 8 prime numbers.
  H = [format(h, '032b') for h in H]

  def rotr(x: str, b: int) -> str:
      b %= 32
      x = x.zfill(32)
      return x[-b:] + x[:-b] if b != 0 else x

  def rsh(x: str, b: int) -> str:
      return format((int(x, 2) >> b) & 0xFFFFFFFF, '032b')

  def xor32(*bins: str) -> str:
      val = 0
      for b in bins:
          val ^= int(b, 2)
      return format(val & 0xFFFFFFFF, '032b')

  def add32(*bins: str) -> str:
      s = sum(int(b, 2) for b in bins) & 0xFFFFFFFF
      return format(s, '032b')

  def sigma0(x):
      return xor32(rotr(x, 7), rotr(x, 18), rsh(x, 3))

  def sigma1(x):
      return xor32(rotr(x, 17), rotr(x, 19), rsh(x, 10))

  def capsig0(x):
      return xor32(rotr(x, 2), rotr(x, 13), rotr(x, 22))

  def capsig1(x):
      return xor32(rotr(x, 6), rotr(x, 11), rotr(x, 25))

  def ch(e, f, g):
      ev = int(e, 2)
      fv = int(f, 2)
      gv = int(g, 2)
      return format(((ev & fv) ^ ((~ev) & gv)) & 0xFFFFFFFF, '032b')

  def Maj(a, b, c):
      av = int(a, 2)
      bv = int(b, 2)
      cv = int(c, 2)
      return format(((av & bv) ^ (av & cv) ^ (bv & cv)) & 0xFFFFFFFF, '032b')

  def wt(wlist, t):
      return add32(sigma1(wlist[t-2]), wlist[t-7], sigma0(wlist[t-15]), wlist[t-16])


  for block in words32:
      w = block[:]
      for t in range(16, 64):
          w.append(wt(w, t))
      a, b, c, d, e, f, g, h = (H[i] for i in range(8))

      for t in range(64):
          T1 = add32(h, capsig1(e), ch(e, f, g), K[t], w[t])
          T2 = add32(capsig0(a), Maj(a, b, c))
          h = g
          g = f
          f = e
          e = add32(d, T1)
          d = c
          c = b
          b = a
          a = add32(T1, T2)

      H[0] = add32(H[0], a)
      H[1] = add32(H[1], b)
      H[2] = add32(H[2], c)
      H[3] = add32(H[3], d)
      H[4] = add32(H[4], e)
      H[5] = add32(H[5], f)
      H[6] = add32(H[6], g)
      H[7] = add32(H[7], h)

  SHA256 = ''.join(format(int(h, 2), '08x') for h in H)
  return SHA256


  #Hey how are you? -> bed99ed8b100b42dcc8cb4876e954afdc60ff49439ae776b9455aa1222ef3ce0
  #I'm fine. Thank you. -> ab9c8394f58d44a20575f1c775ab3e94c61c3c8748fd2d17cf880b23872d172c
  #I am fine. Thank you. -> 82bcac961610fef4b4e488d79fcdbd4d002d42127bcffb2079e95b384e24e228
```
